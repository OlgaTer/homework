#### 1. Переведите содержимое файла purchase_log.txt в словарь purchases вида:
#### {'1840e0b9d4': 'Продукты', ...}  
(содержимое purchase_log.txt помещается в оперативную память компьютера)


```python
f = open('purchase_log.txt', 'r')
context = f.readlines()
f.close()

# удаление строки с заголовками
del(context[0])

# просмотр данных списка
context[:3]
```

```python
# перевод данных в словарь
import json

purchases = {}

for line in context:
    line = line.strip()
    line_json = json.loads(line)
    
    dict_line = list(line_json.values())
    purchases[dict_line[0]] = dict_line[1]
```

#### 2. Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо).
#### Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
(содержимое visit_log.csv не помещается в оперативную память компьютера, используйте только построчную обработку этого файла)

```python
with open('visit_log.csv', 'r') as f_read:
    with open('funnel.csv', 'w') as f_write:
        
        for line in f_read:
            new_line = line.strip().split(',')
            
            if new_line[0] in purchases:
                new_line.append(purchases[new_line[0]])
                
                f_write.write(str(new_line) + '\n')
```
